```python
def translate_to_protein(s):
    protein_seq = "" # Белковая последовательность
    for x in range(0,len(s),3):  # Итерация по РНК последовательности по 3 символа с присваиванием этим символам аминокислот
        codon = s[x:x+3]
        aminoacid = codon_table.get(codon, "")
        if aminoacid == 'Stop':
            break
        protein_seq += aminoacid
    return protein_seq
        
```
